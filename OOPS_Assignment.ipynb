```python
from abc import ABC, abstractmethod

#Vehiclae is a Abstract Class
class Vehicle(ABC):

    def __init__(self, name):
        self.name = name
        self._speed = 0
        self._engine_on = False

    @abstractmethod
    def start(self):
        pass

    def stop(self):
        self._speed = 0
        self._engine_on = False
        print(f"{self.name} stopped.")

```

```python
class Car(Vehicle):

    def __init__(self, name):
        super().__init__(name)
        self.__fuel = 100

    def start(self):
        if self.__fuel > 0:
            self._engine_on = True
            print(f"{self.name} engine started.")
        else:
            print("no fuel. Go to the petrol station.")

    def accelerate(self):
        if self._engine_on:
            self._speed += 20
            self.__fuel -= 5
            print(f"{self.name} Current speed: {self._speed} km/h | fuel: {self.__fuel}")
        else:
            print("Start your engine first.")
```

```python
class Bike(Vehicle):
  def start(self):
    self._engine_on = True
    print(f"{self.name} kick started.")

  def accelerate(self):
    if self._engine_on:
      self._speed += 30
      print(f"{self.name} Current speed: {self._speed} km/h")
    else:
      print(f"Start your {self.name} first")
```

```python
class Bus(Vehicle):

    def __init__(self, name, capacity=40):
        super().__init__(name)
        self.__fuel = 150          # Private fuel attribute
        self.capacity = capacity  # Passenger capacity

    def start(self):
        if self.__fuel > 0:
            self._engine_on = True
            print(f"{self.name} engine started with capacity {self.capacity} passengers.")
        else:
            print(f"{self.name} has no fuel.")

    def accelerate(self):
        if not self._engine_on:
            print("Start the bus engine first.")
            return

        if self.__fuel <= 0:
            print(f"{self.name} ran out of fuel.")
            self.stop()
            return

        # Bus accelerates slower due to size & weight
        self._speed += 10
        self.__fuel -= 8
        print(
            f"{self.name} Current speed: {self._speed} km/h | fuel: {self.__fuel}"
        )

```

```python
def vehicle_simulator(vehicle: Vehicle):
  vehicle.start()
  vehicle.accelerate()
  vehicle.accelerate()
  vehicle.accelerate()
  vehicle.accelerate()
  vehicle.accelerate()
  vehicle.stop()
  vehicle.accelerate()
```

**Calling Classes Objects:**

```python
car = Car("Toyota")
bike = Bike("Yamaha")

vehicle_simulator(car)
print()
vehicle_simulator(bike)
print()
bus = Bus("Volvo Bus", capacity=50)
print()
vehicle_simulator(bus)
```
